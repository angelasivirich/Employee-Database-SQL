```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:Fslowcm12!@localhost:5432/sql_assignment_DB')
```

```python
result = engine.execute("SELECT * FROM employee_salary")

salary_list=[]
for row in result:
    salary_list.append(row["Salary"])
  
plt.hist(salary_list)
plt.title("Distribution Of Salaries")
plt.ylabel("Count")
plt.xlabel("Salary")
plt.show()
```

```python
data = engine.execute("SELECT * FROM average_salary")

title_list=[]
avg_salary_list=[]

for row in data:    
    title_list.append(row["Title"])
    avg_salary_list.append(row["Average Salary"])

plt.bar(title_list, avg_salary_list)
plt.title("Average Salary Per Job Position")
plt.xlabel("Job Position")
plt.ylabel("Average Salary")
plt.show()
```
